Data Set Information:



The dataset includes 244 instances that regroup a data of two regions of Algeria,namely the Bejaia region located in the northeast of Algeria and the Sidi Bel-abbes region located in the northwest of Algeria. 


122 instances for each region.


The period from June 2012 to September 2012. The dataset includes 11 attribues and 1 output attribue (class) The 244 instances have been
classified into fire(138 classes) and not fire (106 classes) classes.


Attributes:

1.  Date:  (DD/MM/YYYY) Day, month (june" to 'september") year (2012) Weather data observations
2. Temp : temperature noon (temperature max) in Celsius degrees 22 to 42
3. RH :  Relative Humidity in %: 21 to 90
4. Ws : Wind speed in km/h: 6 to 29
5. Rain : total day in MMC 0 to 16.8 FWI Components
6. Fine Fuel Moisture Code (FFMC) index from the FWI system: 28.6 to 92.5
7. Duff Moisture Code (DMC) index from the FWI system: 1.1 to 65.9
8. Drought Code (DC) index from the FWI system: 7 to 220.4
9. Initial Spread Index (ISI) index from the FWI system:
10. Buildup Index (BUI) index from the FWI system: 1.1 to 68
11. Fire Weather Index (FWT) Index 0 to 31.1
12. Classes: two classes, namely Fire and not Fire


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

**Data Cleaning**

```python
dataset[dataset.isnull().any(axis=1)]
```

*‘The dataset is converte int two sets based on Region from 122th inde, we can make a new column based on the Region


1: 'Bejaia Region Dataset '


2: 'Sidi-Bel Abbes Region Dataset'
*

```python
dataset.loc[:122,"Region"] = 0
dataset.loc[122:,"Region"] = 1
df = dataset
```

```python
df.info()
```

```python
df.head()
```

```python
df[['Region']] = df[['Region']].astype(int)
```

```python
df.info()
```

```python
df.tail()
```

*Removing the NULL Values*

```python
df = df.dropna().reset_index(drop = True)
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
## remove the 122 row

df = df.drop(122).reset_index(drop=True)
```

```python
df.iloc[122]
```

```python
df.columns
```

```python
## removing blank space from columns names

df.columns = df.columns.str.strip()
df.columns
```

```python

## Change the required columns as integer data type

df[['month','day','year','Temperature','RH','Ws']] = df[['month','day','year','Temperature','RH','Ws']].astype(int)
```

```python
df.info()
```

```python

## CHanging the other columns in float data type

objects = [features for features in df.columns if df[features].dtypes == 'O']
```

```python
objects
```

```python
for i in objects:
    if i!= 'Classes':
        df[i] = df[i].astype(float)
```

```python
df.info()
```

```python
df.head()
```

```python
df.describe()
```

```python
## save the cleaned dataset

df.to_csv('Algerian_forest_fires_cleaned_dataset_.csv',index=False)
```
